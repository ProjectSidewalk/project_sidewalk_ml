```python
from heatmap_depths import depth_generator
from collections import defaultdict
import matplotlib
import matplotlib.pyplot as plt
import math
import numpy as np

matplotlib.rcParams['figure.figsize'] = [12, 10]
```

```python
depths = defaultdict(set)
```

```python
for name, depth in depth_generator(True):
    try:
        depths[name].add(depth)
    except Error as e:
        'Other Error: {}'.format(e)
```

```python
labels = ('Curb Cut', 'Missing Cut', 'Sfc Problem', 'Obstruction',)
```

```python
depths_no_nan = defaultdict(list)
for label in labels:
    for d in depths[label]:
        if not math.isnan(d): depths_no_nan[label].append(d)
```

```python
n_bins = 40

fig, axs = plt.subplots(4, 1, sharey=False, sharex=True, tight_layout=True)

for i, label in enumerate(labels):
    ds = depths_no_nan[label]
    axs[i].hist(ds, bins=n_bins, label=label)
    axs[i].legend()
    
plt.xlabel("Distance (feet?)")

    
```

```python
fig, axs = plt.subplots(2, 1, sharey=False, sharex=True, tight_layout=True)

bins=np.arange(0,35, .5)

axs[0].hist(depths_no_nan['Curb Cut'], bins=bins, label="Curb Cut", color='g')
axs[0].legend()
axs[1].hist(depths_no_nan['Missing Cut'], bins=bins, label="Missing Cut")
axs[1].legend()
plt.xlabel("Distance (feet?)")
```
